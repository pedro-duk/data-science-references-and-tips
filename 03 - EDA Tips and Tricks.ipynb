# Checking for missing data

## Using a heatmap

```python
train = pd.read_csv('data/titanic_train.csv')
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

- Not many values from age are missing, so we can use knowledge from the other columns to fill in;
- A lot of values from cabin are missing, so infering the values would be too imprecise.
  - We could either drop the column or replace it by another feature like a boolean "Cabin Known".
